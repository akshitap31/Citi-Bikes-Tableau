```python
import pandas as pd
import datetime as dt
```

```python
feb_df=pd.read_csv("data/feb2020.csv")
feb_df
```

```python
feb_df["starttime"]=feb_df["starttime"].astype(str).str[:-5]
feb_df["stoptime"]=feb_df["stoptime"].astype(str).str[:-5]
feb_df['startdate'], feb_df['starttime'] = feb_df['starttime'].str.split(' ', 1).str
feb_df['stopdate'], feb_df['stoptime'] = feb_df['stoptime'].str.split(' ', 1).str
feb_df["startdate"]=pd.to_datetime(feb_df["startdate"], format='%Y-%m-%d')
feb_df=feb_df.loc[feb_df["startdate"]> "2020-2-20"]
feb_df
```

```python
feb_df.to_csv("db/feb.csv")
```

```python
apr_df=pd.read_csv("data/apr2020.csv")
apr_df
```

```python
apr_df["starttime"]=apr_df["starttime"].astype(str).str[:-5]
apr_df["stoptime"]=apr_df["stoptime"].astype(str).str[:-5]
apr_df['startdate'], apr_df['starttime'] = apr_df['starttime'].str.split(' ', 1).str
apr_df['stopdate'], apr_df['stoptime'] = apr_df['stoptime'].str.split(' ', 1).str
apr_df["startdate"]=pd.to_datetime(apr_df["startdate"], format='%Y-%m-%d')
apr_df=apr_df.loc[apr_df["startdate"]> "2020-4-20"]
apr_df
```

```python
apr_df.to_csv("db/apr.csv")
```

```python
june_df=pd.read_csv("data/june2020.csv")
june_df
```

```python
june_df["starttime"]=june_df["starttime"].astype(str).str[:-5]
june_df["stoptime"]=june_df["stoptime"].astype(str).str[:-5]
june_df['startdate'], june_df['starttime'] = june_df['starttime'].str.split(' ', 1).str
june_df['stopdate'], june_df['stoptime'] = june_df['stoptime'].str.split(' ', 1).str
june_df["startdate"]=pd.to_datetime(june_df["startdate"], format='%Y-%m-%d')
june_df=june_df.loc[june_df["startdate"]> "2020-6-20"]
june_df
```

```python
june_df.to_csv("db/june.csv")
```

```python
feb_df["starthour"]=feb_df["starttime"].astype(str).str[:-6]
apr_df["starthour"]=apr_df["starttime"].astype(str).str[:-6]
june_df["starthour"]=june_df["starttime"].astype(str).str[:-6]
feb_df["endhour"]=feb_df["stoptime"].astype(str).str[:-6]
apr_df["endhour"]=apr_df["stoptime"].astype(str).str[:-6]
june_df["endhour"]=june_df["stoptime"].astype(str).str[:-6]
```

```python
new_df=pd.DataFrame(apr_df.groupby(["starthour"]).count())
new_df["starthour"]=new_df["tripduration"]+feb_df.groupby(["starthour"]).count()['starthour']
new_df["starthour"]=new_df["tripduration"]+apr_df.groupby(["starthour"]).count()['starthour']
new_df
```

```python
new=pd.DataFrame(apr_df.groupby(["endhour"]).count())
new
```

```python
new_df["stophour"]=feb_df.groupby(["endhour"]).count()['tripduration']
new_df["stophour"]=new_df["stophour"]+ apr_df.groupby(["endhour"]).count()['tripduration']
new_df["stophour"]=new_df["stophour"]+june_df.groupby(["endhour"]).count()['tripduration']
new_df
```

```python
new_df.to_csv("db/hours.csv")
```
